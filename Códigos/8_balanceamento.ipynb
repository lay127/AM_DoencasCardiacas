## Separação dos valores de treino e teste 

```python
X_prev = base.drop(columns=['Doença_cardíaca'])
y_classe = base['Doença_cardíaca']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X_prev, y_classe, test_size=0.3, random_state=42)
```

```python
np.savez('../Bases/Criados/Treino_Teste/treino_teste.npz', X_treino=X_treino, X_teste=X_teste, y_treino=y_treino, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_treino_original = pd.concat([X_treino.reset_index(drop=True), y_treino.reset_index(drop=True)], axis=1)
df_teste_original = pd.concat([X_teste.reset_index(drop=True), y_teste.reset_index(drop=True)], axis=1)

df_treino_original.to_csv("../Bases/Criados/Treino_Teste/treino.csv", index=False)
df_teste_original.to_csv("../Bases/Criados/Treino_Teste/teste.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define o tamanho da imagem em HD
plt.figure(figsize=(20, 15))  # (largura, altura) em polegadas

# Gera a matriz de correlação
correlacao = base.corr()

# Cria o heatmap com melhorias
sns.heatmap(correlacao, 
            annot=False,         # Se quiser mostrar os valores, mude para True
            cmap='coolwarm',     # Paleta de cores
            cbar=True,           # Exibe a barra de cores
            xticklabels=True, 
            yticklabels=True)

# Rotaciona os nomes das colunas e linhas para caber melhor
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)

# Título opcional
plt.title('Matriz de Correlação', fontsize=16)

# Ajusta layout para evitar corte de texto
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

## Random UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy='not minority', random_state=42)
X_resampled, y_resampled = undersample.fit_resample(X_treino, y_treino)
```

```python
np.savez('../Bases/Criados/Treino_Teste/random.npz', X_treino=X_resampled, X_teste=X_teste, y_treino=y_resampled, y_teste=y_teste, feature_names=X_treino.columns)
```

```python
df_random = pd.concat([pd.DataFrame(X_resampled, columns=X_treino.columns),
                       pd.DataFrame(y_resampled, columns=['Q06306'])], axis=1)
df_random.to_csv("../Bases/Criados/Treino_Teste/random.csv", index=False)
```
